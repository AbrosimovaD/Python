```python
import requests
```

```python
countries = ['london', 'svo', 'череповец']
```

```python
url = 'https://wttr.in'
```

```python
def get_weather(countries):
    params = {
        'lang' : 'ru',
        'M':'',
        'm':'',
        'n':'',
        'q':'',
        'T':''
    }
    forecast = ''
    for place in countries:
        response = requests.get(url+'/'+place, params = params)
        response.raise_for_status()
        forecast += response.text
    return print(forecast)
```

```python
get_weather(countries)
```
