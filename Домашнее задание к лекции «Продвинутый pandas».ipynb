#### Задание 3
Используйте файл с оценками фильмов *ml-latest-small/ratings.csv*. 

Посчитайте **среднее время жизни пользователей, которые выставили более 100 оценок**.

*Под временем жизни понимается разница между максимальным и минимальным значениями столбца timestamp для данного значения userId.*

```python
ratings = pd.read_csv('D://ml-latest-small/ratings.csv')
```

```python
ratings.head()
```

```python
ratings.groupby('userId').agg('count').reset_index()[['userId', 'timestamp']]
```

```python
def mean_life(data):
    return data.timestamp.max() -  data.timestamp.min()
```

```python
#среднее время жизни пользователей, которые выставили более 100 оценок
ratings[ratings.userId.isin(ratings.groupby('userId').agg('count').reset_index()['userId'])].groupby('userId').apply(mean_life).reset_index().rename(columns = {0: 'mean_life'})
```

```python
ratings_update = ratings.merge(ratings[ratings.userId.isin(ratings.groupby('userId').agg('count').reset_index()['userId'])].groupby('userId').apply(mean_life).reset_index().rename(columns = {0: 'mean_life'}), how = 'left', on = 'userId')
```
